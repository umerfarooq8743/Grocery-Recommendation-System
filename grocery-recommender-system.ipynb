## Popularity Based Recommender System

```python
ratings_with_name = ratings.merge(items,on='ISBN')
ratings_with_name.sample(10)
```

```python
num_rating_df = ratings_with_name.groupby('Grocery-Title').count()['Item-Rating'].reset_index()
num_rating_df.rename(columns={'Item-Rating':'num_ratings'},inplace=True)
num_rating_df.sample(10)
```

```python
avg_rating_df = ratings_with_name.groupby('Grocery-Title').mean()['Item-Rating'].reset_index()
avg_rating_df.rename(columns={'Item-Rating':'avg_rating'},inplace=True)
avg_rating_df.sample(10)
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Grocery-Title')
popular_df.sample(10)
```

```python
popular_df = popular_df[popular_df['num_ratings']>=30].sort_values('avg_rating',ascending=False).head(50)
popular_df.sample(10)
```

```python
popular_df = popular_df.merge(items,on='Grocery-Title').drop_duplicates('Grocery-Title')[['Grocery-Title','Book-Price','Image-URL-M','num_ratings','avg_rating']]
popular_df.sample(10)
```

## Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Item-Rating'] > 2
regular_users = x[x].index
regular_users
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(regular_users)]
filtered_rating.sample(10)
```

```python
y = filtered_rating.groupby('Grocery-Title').count()['Item-Rating']>=1
famous_items = y[y].index
famous_items
```

```python
final_ratings = filtered_rating[filtered_rating['Grocery-Title'].isin(famous_items)]
final_ratings.sample(10)
```

```python
pt = final_ratings.pivot_table(index='Grocery-Title',columns='User-ID',values='Item-Rating')
pt.sample(10)
```

```python
pt.fillna(0,inplace=True)
```

```python
pt.sample(20)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
similarity_scores
```

```python
similarity_scores.shape
```

```python
def recommend(product_name):
    # index fetch
    index = np.where(pt.index==product_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[0:14]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = items[items['Grocery-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Grocery-Title')['Grocery-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Grocery-Title')['Book-Price'].values))
        item.extend(list(temp_df.drop_duplicates('Grocery-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('dew')
```
